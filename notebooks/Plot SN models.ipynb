```python
%load_ext autoreload
%autoreload 2
import astropy.units as u

from astroduet import models
sims = models.Simulations()

import numpy as np
from matplotlib import pyplot as plt
from astropy.visualization import quantity_support
import matplotlib
font = {'family' : 'sans',
        'weight' : 'bold',
        'size'   : 18}

matplotlib.rc('font', **font)

dist_check = 300 * u.Mpc
early = 1
late = 7
```

```python
#sims.parse_sne_rsg()
#sims.parse_sne_bsg()
#sims.parse_sne_ysg()

```

```python
plt.figure()
for rsg in sims.sne_rsg_simulations:
    print(rsg)
    fits_file = models.fits_file(rsg)
#    print(fits_file)
    
    rsg_model = models.load_model_ABmag(fits_file, dist=dist_check)
    days = rsg_model[0].to(u.d)


    plt.plot(days, rsg_model[1])
    
    in_range = np.where((days.value > early) & (days.value < late))
    print(rsg_model[1][in_range].mean())

plt.xlim(0.01, 21)
plt.ylim([25, 19])
plt.xlabel('Days')
plt.ylabel('Model DUET1 mag')
plt.xscale('Log')
plt.figure()
for rsg in sims.sne_rsg_simulations:
#    print(rsg)
    fits_file = models.fits_file(rsg)
#    print(fits_file)
    
    rsg_model = models.load_model_ABmag(fits_file, dist=dist_check)
    days = rsg_model[0].to(u.d)


    plt.plot(days, rsg_model[2])
    
    in_range = np.where((days.value > early) & (days.value < late))
#    print(rsg_model[2][in_range].mean())
    
    
    
plt.xlim(0.01, 21)
plt.ylim([25, 19])
plt.xlabel('Days')
plt.ylabel('Model DUET2 mag')
plt.xscale('Log')
plt.show()
```
